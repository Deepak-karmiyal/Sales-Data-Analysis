Sales Data Analysis: Extracting Insights to Drive Business Growth


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/DEEPAK SINGH/Diwali Sales Data.csv",encoding = 'unicode_escape')
```

```python
df = data.copy()
```

```python
df.head()
```

```python
df.shape
```

```python
data.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1,inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df["Amount"].astype('int')
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df['Marital_Status']= df['Marital_Status'].astype('str')
```

```python
df['Marital_Status']= df['Marital_Status'].str.replace('0','married').astype('O')
```

```python
df['Marital_Status']= df['Marital_Status'].str.replace('1','unmarried').astype('O')
```

```python
df.head()
```

```python
df.describe()
```

Gender:
find the most no. of buyers from the gender category

```python
df['Gender'].value_counts()
```

```python
gen = sns.countplot(x = 'Gender',data = df,hue = 'Gender')
for bars in gen.containers:
    gen.bar_label(bars)

plt.show()
```

from the data below we can see that most of the buyers are females and even the purchasing power of female is more as compare then male

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen,hue = 'Gender',legend=False)
```

Age:find which age category is buying 

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')


for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

from this data we can clearly see that most number of buyers are the age group of 26-25 and the maximum no. of buyers are females

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age)

for bars in ax.containers:
    ax.bar_label(bars)
```

STATE:
FIND WHICH STATE ARE BUYING MORE PRODUCT

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x = 'State',y = 'Orders',palette='Set2',hue='State')
```

FROM THE DATA MOST OF BUYERS ARE FROM UTTAR PRADESH,MAHARASHTRA,KARNATAKA AND DELHI AND WE CAN SEE THAT TOTAL MAXIMUM ORDER ARE ALSO FROM THERE STATES

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x = 'State',y = 'Amount',palette='Set2',hue='State')
```

marital status

```python
ax = sns.countplot(data = df,x = 'Marital_Status',palette='Set2',hue='Marital_Status',legend=False)

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```


from the above graphs we can see that most od buyer are married (women) and they have high purchasing power


Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df,x = 'Occupation',palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

product category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df,x = 'Product_Category',palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```
